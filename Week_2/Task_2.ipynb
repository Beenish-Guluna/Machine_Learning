### Mean Imputation

```python


df1 = df.copy()

df1["Glucose"] = df1["Glucose"].fillna(df1["Glucose"].mean())
df1["Serum_Insulin"] = df1["Serum_Insulin"].fillna(df1["Serum_Insulin"].mean())
df1["Diastolic_BP"] = df1["Diastolic_BP"].fillna(df1["Diastolic_BP"].mean())
df1["Skin_Fold"] = df1["Skin_Fold"].fillna(df1["Skin_Fold"].mean())
df1["BMI"] = df1["BMI"].fillna(df1["BMI"].mean())

df1.head()




```

### Median Imputation

```python
df2 = df.copy()


df2["Glucose"] = df2["Glucose"].fillna(df2["Glucose"].median())

df2["Serum_Insulin"] = df2["Serum_Insulin"].fillna(df2["Serum_Insulin"].median())

df2["Diastolic_BP"] = df2["Diastolic_BP"].fillna(df2["Diastolic_BP"].median())

df2["Skin_Fold"] = df2["Skin_Fold"].fillna(df2["Skin_Fold"].median())

df2["BMI"] = df2["BMI"].fillna(df2["BMI"].median())



df2.head()
```

### Mode Imputation

```python
df3 = df.copy()


columns_to_fill = ["Glucose", "Serum_Insulin", "Diastolic_BP", "Skin_Fold", "BMI"]

for col in columns_to_fill:
    df3[col] = df3[col].fillna(df3[col].mode())




df3.head()
```

Normalize the dataset using Min-Max Scaling.

```python

# copy the data 
df_min_max_scaled = df.copy() 
  
# apply normalization techniques 
for column in df_min_max_scaled.columns: 
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())     
  
# view normalized data 
display(df_min_max_scaled)
```

### Standardize features using z-score scaling.

```python
df_ZScore = df1.copy()

for column in df_ZScore.columns:
    if column in ["Pregnant", "Age", "Class"]:
        continue
    else:
        df_ZScore[column] = (df_ZScore[column] - df_ZScore[column].mean()) / df_ZScore[column].std()

display(df_ZScore)
```

One_Hot Encoding

```python
dt = pd.read_csv("autism.csv")
dt.head()
```

```python
dt = pd.get_dummies(dt, columns=["Result"])
dt.head()
```

Label Encoding

```python
import sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
encoder = LabelEncoder()

# Apply label encoding on 'Functioning' column
dt['Functioning_encoded'] = encoder.fit_transform(dt['Functioning'])

# Check the result
display(dt)
```

Plot correlations using a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = dt.select_dtypes(include=["number"])

# Calculate correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
