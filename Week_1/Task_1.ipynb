```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Load the data
data = pd.read_csv("thyroid_cancer_risk_data.csv")

# Type of data present in the file
type(data)
print(data.dtypes)
```

```python
# Display first 5 elements of the data

df = data.head(500)
df.head()
```

```python
# PLotting scatter plot with seaborn

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df, 
    x='TSH_Level', 
    y='Nodule_Size', 
    hue='Thyroid_Cancer_Risk', 
    alpha=0.7
)

plt.title('TSH Level vs Nodule Size (First 500 Entries)')
plt.xlabel('TSH Level')
plt.ylabel('Nodule Size')
plt.legend(title='Thyroid Cancer Risk')
plt.show()

plt.show()
```
